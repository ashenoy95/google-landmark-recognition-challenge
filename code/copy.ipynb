```python
import os
from tqdm import tqdm
from random import shuffle
from shutil import copy2
```

```python
path = '/share/jproject/fg538/r-006-gpu-4/data'

for folder in tqdm(os.listdir('{}/tensorflow-image-classifier/training_dataset'.format(path))):
    file_count = 0
    filenames = os.listdir('{}/tensorflow-image-classifier/training_dataset/{}/'.format(path, folder))[:10000]
    shuffle(filenames)
    for file in filenames:
        filename_source = '{}/tensorflow-image-classifier/training_dataset/{}/{}'.format(path, folder, file)
        if file_count <= 2000:
            dirname = '{}/keras_data/val_dir/{}/'.format(path, folder)
            os.makedirs(os.path.dirname(dirname), exist_ok=True)
            copy2(filename_source, dirname)
        else:
            dirname = '{}/keras_data/train_dir/{}/'.format(path, folder)
            os.makedirs(os.path.dirname(dirname), exist_ok=True)
            copy2(filename_source, dirname)
        file_count+=1
```

```python
path = '/share/jproject/fg538/r-006-gpu-4/data'
train_count = {}
val_count = {}

for folder in tqdm(os.listdir('{}/keras_data/train_dir/'.format(path))):
    train_count[folder] = len(os.listdir('{}/keras_data/train_dir/{}/'.format(path, folder)))
    val_count[folder] = len(os.listdir('{}/keras_data/val_dir/{}/'.format(path, folder)))
```
